```python
# custom function for neat plotting
def apply_custom_theme(fig, title, xaxis_title, yaxis_title):
    fig.update_layout(
        title=title,
        xaxis_title=xaxis_title,
        yaxis_title=yaxis_title,
        font=dict(
            family="Arial, sans-serif",
            size=12,
            color="black"
        ),
        template="plotly_white",
        margin=dict(l=50, r=50, t=50, b=50),
        width=600,
        height=400
    )

    if isinstance(fig.data[0], go.Bar):
        fig.update_traces(marker_color='royalblue')
```

```python
# visualising NaNs

null_values = df.isnull()

fig = px.imshow(null_values,
                labels=dict(x="Columns", y="Rows", color="Missing Value"),
                x=null_values.columns,
                y=null_values.index,
                title='Heatmap of Missing Values',
                color_continuous_scale=[[0, 'white'], [1, 'royalblue']])

fig.update_layout(
    margin=dict(l=50, r=50, t=50, b=50),
)
```

```python
# clinical trials started by year

df['Start Date'] = pd.to_datetime(df['Start Date'])
df['Year'] = df['Start Date'].dt.year

trials_per_year = df['Year'].value_counts().sort_index().reset_index()
trials_per_year.columns = ['Year', 'Count']

fig = px.bar(trials_per_year,
             x='Year',
             y='Count',
             text='Count',
             title='Number of Clinical Trials Started by Year',
             labels={'Year': 'Year', 'Count': 'Number of Trials'})

apply_custom_theme(fig,
                   title='Number of Clinical Trials Started by Year',
                   xaxis_title='Year',
                   yaxis_title='Number of Trials')

fig.update_xaxes(tickangle=45, dtick=1)
fig.show()
```

```python
# status distribution among clinical trials

status_counts = df['Study Status'].value_counts().reset_index()
status_counts.columns = ['Status', 'Count']
status_counts['Percentage'] = (status_counts['Count'] / status_counts['Count'].sum()) * 100

status_counts = status_counts.sort_values(by='Count', ascending=True)

fig = px.bar(status_counts,
             y='Status',
             x='Count',
             text=status_counts['Percentage'].apply(lambda x: f'{x:.2f}%'),
             title='Number of Clinical Trials by Overall Status',
             labels={'Status': 'Overall Status', 'Count': 'Number of Trials'},
             orientation='h')

apply_custom_theme(fig,
                   title='Number of Clinical Trials by Overall Status',
                   xaxis_title='Number of Trials',
                   yaxis_title='Overall Status')

fig.show()
```

```python
# phases vs enrollement along with violin plot to check for outliers

fig = px.violin(df,
             y='Phases',
             x='Enrollment',
             title='Enrollment Numbers by Phase',
             labels={'Phases': 'Phase', 'Enrollment': 'Enrollment'},
             orientation='h')

apply_custom_theme(fig,
                   title='Enrollment Numbers by Phase',
                   xaxis_title='Enrollment',
                   yaxis_title='Phase')

fig.update_xaxes(tickvals=[0, 100000, 200000, 300000, 400000, 500000],
                 ticktext=['0', '100k', '200k', '300k', '400k', '500k'])

fig.show()
```

```python
# making plots of frequency for all the categorical columns

categorical_columns = ['Study Results', 'Sex', 'Age', 'Phases', 'Funder Type']

for column in categorical_columns:
    freq = df[column].value_counts().reset_index()
    freq.columns = [column, 'Count']

    fig = px.bar(freq,
                    y=column,
                    x='Count',
                    title=f'Frequency of {column}',
                    labels={column: column, 'Count': 'Number of Trials'},
                    orientation='h')

    apply_custom_theme(fig,
                        title=f'Frequency of {column}',
                        xaxis_title='Number of Trials',
                        yaxis_title=column)

    fig.show()
    print('\n')
```

```python
# violin plot to visualise the outliers in enrollment numbers

fig = px.violin(df,
                y='Enrollment',
                title='Distribution of Enrollment Numbers',
                labels={'Enrollment': 'Enrollment'})

apply_custom_theme(fig,
                    title='Distribution of Enrollment Numbers',
                    xaxis_title='',
                    yaxis_title='Enrollment')

fig.show()
```
