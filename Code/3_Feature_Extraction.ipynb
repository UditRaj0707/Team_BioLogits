## Fetch Cities


```python
import requests
```

```python
def fetch_cities(nct_id):
    base_url = f"https://clinicaltrials.gov/api/v2/studies/{nct_id}"
    params = {
        "fields": "LocationCity",  # Specify fields to fetch
        "format": "json",  # Default format
    }
    headers = {"accept": "application/json"}
    response = requests.get(base_url, params=params, headers=headers)
    if response.status_code == 200:
        data = response.json()
        try:
          locs = data['protocolSection']['contactsLocationsModule']['locations']
          cities = []
          for loc in locs:
              cities.append(loc['city'])
          return cities
        except:
          return None
    else:
        return f"Failed to fetch data. Status code: {response.status_code}"
```

## Population

```python
popu_data = pd.read_csv("/content/drive/MyDrive/NOVARTIS NEST - BioLogits/Data/worldcities.csv")
```

```python
popu_data.head()
```

```python
from tqdm import tqdm
```

```python
data = city_df.copy()
```

```python
not_found_cities = []
for i, cities in enumerate(tqdm(data['Cities'])):
  popu = 0
  for city in cities:
    try:
      popu += popu_data[popu_data['city_ascii'] == city]['population'].values[0]
    except:
      not_found_cities.append(city)
      continue
  data.at[i, 'population'] = popu
```

```python
from unidecode import unidecode
def convert_cities_to_ascii(city_list):
    return [unidecode(city) for city in city_list]

# Apply the function to the 'Cities' column
data["Cities_ASCII"] = data["Cities"].apply(convert_cities_to_ascii)
```

```python
from tqdm import tqdm

# Convert `popu_data` to a dictionary for faster lookups
popu_dict = popu_data.set_index('city_ascii')['population'].to_dict()
```

```python


# Initialize a population column and not_found_cities list
data['population'] = 0
not_found_cities = []

# Process cities efficiently
def calculate_population(cities):
    popu = 0
    for city in cities:
        if city in popu_dict:
            popu += popu_dict[city]
        else:
            not_found_cities.append(city)
    return popu


```

```python
data['Cities'][0]
```

```python
data['Cities'] = data['Cities'].apply(lambda x: eval(x))
```

```python
# Apply the function to calculate population for all rows
data['population'] = tqdm(data['Cities_ASCII'].apply(calculate_population))
```
