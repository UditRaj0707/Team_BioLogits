## Calculate Entropy of Eligibilty Criteria

```python
import math
from collections import Counter

def calculate_shannon_entropy(text):
    """
    Calculate the Shannon entropy (1st-order entropy) of a string.

    Args:
        text (str): The input string.

    Returns:
        float: Shannon entropy of the string.
    """
    # Count the frequency of each unique character
    char_counts = Counter(text)
    total_chars = len(text)

    # Calculate the probabilities of each character
    probabilities = [count / total_chars for count in char_counts.values()]

    # Calculate Shannon entropy
    shannon_entropy = -sum(p * math.log2(p) for p in probabilities if p > 0)
    return shannon_entropy

def calculate_second_order_entropy(text):
    """
    Calculate the second-order entropy of a string.

    Args:
        text (str): The input string.

    Returns:
        float: Second-order entropy of the string.
    """
    # Generate all adjacent character pairs (bigrams)
    bigrams = [text[i:i+2] for i in range(len(text) - 1)]

    # Count the frequency of each unique bigram
    bigram_counts = Counter(bigrams)
    total_bigrams = len(bigrams)

    # Calculate the probabilities of each bigram
    probabilities = [count / total_bigrams for count in bigram_counts.values()]

    # Calculate second-order entropy
    second_order_entropy = -sum(p * math.log2(p) for p in probabilities if p > 0)
    return second_order_entropy
```

```python
from tqdm import tqdm
```

```python
for i, ec in enumerate(tqdm(data['Eligibility Criteria'])):
  data.loc[i, 'ec_entropy_1st'] = calculate_shannon_entropy(ec)
  data.loc[i, 'ec_entropy_2nd'] = calculate_second_order_entropy(ec)
```

```python
import pandas as pd

# Load the dataset
ghe_data_path = "/content/drive/MyDrive/NOVARTIS NEST - BioLogits/Data/GHE_Data - All ages (1).csv"  # Replace with the actual file path
ghe_data = pd.read_csv(ghe_data_path)

```

```python
ghe_data = ghe_data[4:]
ghe_data.reset_index(drop=True, inplace=True)
```

```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
```

```python
ghe_data_df = ghe_data.copy()
ghe_causes = ghe_data_df["GHE cause"].dropna().unique()  # Extract unique GHE causes
```

```python
ghe_causes_final = ['Communicable, maternal, perinatal and nutritional conditions', 'Noncommunicable diseases', 'Injuries']
```

```python



conditions = country_list_df["Conditions"].unique()
ghe_causes = ghe_causes_final

model = SentenceTransformer('abhinand/MedEmbed-small-v0.1')

conditions_embeddings = model.encode(conditions, convert_to_tensor=True)
ghe_causes_embeddings = model.encode(ghe_causes, convert_to_tensor=True)

cosine_scores = util.cos_sim(conditions_embeddings, ghe_causes_embeddings)

threshold = 0.7

matches = []
for condition_idx, condition in enumerate(conditions):
    for cause_idx, ghe_cause in enumerate(ghe_causes):
        score = cosine_scores[condition_idx][cause_idx].item()
        if score >= threshold:
            matches.append({
                "Condition": condition,
                "GHE cause": ghe_cause,
                "Similarity": score
            })

matches_df = pd.DataFrame(matches)

matches_df.to_csv("condition_ghe_cause_mapping.csv", index=False)

print(matches_df.sort_values(by="Similarity", ascending=False).head(20))
```

```python
def get_top_mappings(df, similarity_col):
    return (
        df.sort_values(["Condition", similarity_col], ascending=[True, False])
        .drop_duplicates(subset="Condition", keep="first")
    )

# Process each DataFrame to keep only the top similarity per Condition
dc_1_top = get_top_mappings(dc_1_mapping, "Similarity_dc_1")
dc_2_top = get_top_mappings(dc_2_mapping, "Similarity_dc_2")
disease_top = get_top_mappings(disease_mapping, "Similarity_disease")
ghe_top = get_top_mappings(ghe_cause_mapping, "Similarity_ghe_cause")

# Merge all top mappings into one DataFrame
direct_matches = (
    dc_1_top.merge(dc_2_top, on="Condition", how="outer")
    .merge(disease_top, on="Condition", how="outer")
    .merge(ghe_top, on="Condition", how="outer")
)
```

```python
col_map = {
    "Similarity_dc_1": "DC_1",
    "Similarity_dc_2": "DC_2",
    "Similarity_disease": "Disease",
    "Similarity_ghe_cause": "GHE cause",
}
```

```python
for i, condition in enumerate(direct_matches['Condition']):
  max_sim = (0, "")
  for col, col_cause in col_map.items():
    if pd.isna(direct_matches.loc[i, col]):
      continue
    if direct_matches.loc[i, col] > max_sim[0]:
      max_sim = direct_matches.loc[i, col], col_cause
  direct_matches.loc[i, 'max_similarity'] = max_sim[0]
  direct_matches.loc[i, 'max_similarity_col'] = direct_matches.loc[i, max_sim[1]]
```
