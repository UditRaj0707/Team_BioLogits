```python
# remove the row with max Enrollment value
final_df = final_df[final_df['Enrollment'] != final_df['Enrollment'].max()]
```

```python
nan_list = ['Allocation_NA', 'Allocation_NON_RANDOMIZED',
       'Allocation_RANDOMIZED', 'Intervention Model_CROSSOVER',
       'Intervention Model_PARALLEL', 'Intervention Model_SEQUENTIAL',
       'Intervention Model_SINGLE_GROUP', 'Masking Complexity_DOUBLE',
       'Masking Complexity_NONE', 'Masking Complexity_QUADRUPLE',
       'Masking Complexity_SINGLE', 'Masking Complexity_TRIPLE',
       'Primary Purpose_BASIC_SCIENCE', 'Primary Purpose_OTHER',
       'Primary Purpose_PREVENTION', 'Primary Purpose_TREATMENT']

final_df[nan_list] = final_df[nan_list].fillna(0)
```

```python
columns_to_fill = ['sponsor_success_rate', 'prevalent_disease', 'population']
final_df[columns_to_fill] = final_df[columns_to_fill].apply(lambda col: col.fillna(col.mean()))
```

```python
final_df.isna().sum()
```

```python
final_df.columns
```

```python
final_df.describe()
```

```python
numerical_cols = ['Enrollment', 'ec_entropy_1st', 'ec_entropy_2nd', 'sponsor_success_rate',
       'prevalent_disease', 'no_of_sites', 'population',
       'competition']

# Apply Min-Max Scaling
final_df[numerical_cols] = final_df[numerical_cols].apply(lambda col: (col - col.min()) / (col.max() - col.min()))
```

```python
final_df.describe()
```

```python
final_df['mesh_id'] = final_df['mesh_id'].apply(lambda x: eval(x))
```

```python
freq_dict = {}
for sub_list in final_df['mesh_id']:
  for item in sub_list:
    if item in freq_dict:
      freq_dict[item] += 1
    else:
      freq_dict[item] = 1
```

```python
# Frequency encoding: Replace each list with the sum of frequencies of its elements
for index, row in final_df.iterrows():
    freq_sum = sum(freq_dict[item] for item in row['mesh_id'])/len(row['mesh_id'])
    final_df.at[index, 'mesh_id_encoded'] = freq_sum
```

```python
final_df['mesh_id_encoded']
```

```python
numerical_cols = ['mesh_id_encoded']

# Apply Min-Max Scaling
final_df[numerical_cols] = final_df[numerical_cols].apply(lambda col: (col - col.min()) / (col.max() - col.min()))
```

```python
final_dataset = final_df.drop('mesh_id', axis=1)
```

```python
one_hot_cols = ['Age']

final_dataset = pd.get_dummies(final_dataset, columns=one_hot_cols, dtype=int)
```
